#Question no. 3

#Question no. 5

5.	Create a table giving the visibility of ‘distributorId’ wise, the count of orders in different status and the percentage of status_b (status) of total orders as shown below
Output Format
image.png


```python
data_3.head(1)
```

```python
z1 = data_3.groupby('distributorId')['status'].apply(lambda x: (x=='status_a').sum()).reset_index()
```

```python
z2 = data_3.groupby('distributorId')['status'].apply(lambda x: (x=='status_b').sum()).reset_index()
```

```python
z = data_3.groupby(['distributorId'])['orderId'].count().reset_index()
```

```python
z['Status_a (count)'] = z1['status']
z['Status_b (count)'] = z2['status']
z['Status_b (percentage)'] = z['Status_b (count)']*100/z['orderId']
z.rename(columns = {'orderId':'totalOrders'}, inplace = True)
z
```

```python
q5= z
csv_q5 = q5.to_csv('5.csv')
```
